## **Part A** - Creating a file
---

Here we will be retrieving data from the user input, adding it to a list, and then publishing it to a file so that it can persist after the program has been exited.<br/><br/>

### Your Task

Create a basic loop that reads in scores from the user until the user enters the sentinel value of -1. Each score should be added to a list. <br/><br/>

After the loop has been exited, the program should take the top 10 scores and print them in order (highest to lowest) to the screen, and also to a file called `highscores.txt` and then the program should exit.

<br/>


### Your Answer

```python
# Student Name: PAUL NICOLLI
# Student Number: 1037974

scores = []

# Read scores until sentinel value -1 is entered
score = int(input("Enter a score (-1 to stop): "))

while score != -1:
    scores.append(score)
    score = int(input("Enter a score (-1 to stop): "))


# Sort scores from highest to lowest
scores.sort(reverse=True)

# Take the top 10 scores
top_scores = scores[:10]


# Print top scores to screen
print()
print("Top 10 High Scores")
print("------------------")

for score in top_scores:
    print(score)


# Write top scores to file
with open("highscores.txt", "w") as file:
    for score in top_scores:
        file.write(str(score) + "\n")


print()
print("High scores saved to highscores.txt")
```

```python
# Student Name: PAUL NICOLLI
# Student Number: 1037974

# Completed functionality:
# [x] Save and display player name against score
# [x] Automatically save and display date/time of highscore
# [x] Menu-based program
# [x] Add new score
# [x] View scores
# [x] Delete score
# [x] Modify score
# [x] Clear highscore list
# [x] Exit and save

# Further improvement ideas:
# - Add input validation so scores cannot be negative except for sentinel values
# - Add difficulty levels or game names against each score
# - Add search by player name
# - Add a maximum highscore limit chosen by the user


from datetime import datetime


highscores = []


# Load existing scores from highscores.txt
def load_scores():
    try:
        with open("highscores.txt", "r") as file:
            for line in file:
                line = line.strip()

                if line == "":
                    continue

                data = line.split(",")

                try:
                    # Reads the newer format: name, score, date
                    if len(data) >= 3:
                        name = data[0]
                        score = int(data[1])
                        date_time = data[2]

                        highscores.append({
                            "name": name,
                            "score": score,
                            "date": date_time
                        })

                    # Reads the older format from Part A/B: score only
                    elif len(data) == 1:
                        score = int(data[0])

                        highscores.append({
                            "name": "Unknown",
                            "score": score,
                            "date": "No date saved"
                        })

                except ValueError:
                    # Skip invalid lines
                    pass

    except FileNotFoundError:
        print("No highscores.txt file found. A new one will be created.")


# Save the top 10 scores to the file
def save_scores():
    highscores.sort(key=lambda item: item["score"], reverse=True)

    top_scores = highscores[:10]

    with open("highscores.txt", "w") as file:
        for item in top_scores:
            file.write(f"{item['name']},{item['score']},{item['date']}\n")


# Display the current top 10 high scores
def view_scores():
    highscores.sort(key=lambda item: item["score"], reverse=True)

    print()
    print("High Scores")
    print("-----------")

    if len(highscores) == 0:
        print("No high scores found.")
    else:
        top_scores = highscores[:10]

        for index, item in enumerate(top_scores, start=1):
            print(f"{index}. {item['name']} - {item['score']} - {item['date']}")

    print()


# Add a new score with the current date and time
def add_score():
    name = input("Enter player name: ")
    score = int(input("Enter score: "))

    date_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    highscores.append({
        "name": name,
        "score": score,
        "date": date_time
    })

    print("Score added.")


# Delete a selected score
def delete_score():
    view_scores()

    if len(highscores) == 0:
        return

    number = int(input("Enter the score number to delete: "))

    highscores.sort(key=lambda item: item["score"], reverse=True)

    if number >= 1 and number <= len(highscores):
        removed_score = highscores.pop(number - 1)
        print(f"Deleted score for {removed_score['name']}.")
    else:
        print("Invalid score number.")


# Modify a selected score
def modify_score():
    view_scores()

    if len(highscores) == 0:
        return

    number = int(input("Enter the score number to modify: "))

    highscores.sort(key=lambda item: item["score"], reverse=True)

    if number >= 1 and number <= len(highscores):
        item = highscores[number - 1]

        new_name = input("Enter new player name: ")
        new_score = int(input("Enter new score: "))
        new_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        item["name"] = new_name
        item["score"] = new_score
        item["date"] = new_date

        print("Score modified.")
    else:
        print("Invalid score number.")


# Clear all saved scores
def clear_scores():
    confirm = input("Are you sure you want to clear all scores? yes/no: ")

    if confirm.lower() == "yes":
        highscores.clear()
        print("Highscore list cleared.")
    else:
        print("Clear cancelled.")


# Display the menu options
def show_menu():
    print()
    print("Highscore Menu")
    print("--------------")
    print("1. Add new score")
    print("2. View scores")
    print("3. Delete score")
    print("4. Modify score")
    print("5. Clear highscore list")
    print("6. Exit and save")


# Main program
load_scores()

choice = ""

while choice != "6":
    show_menu()
    choice = input("Enter your choice: ")

    if choice == "1":
        add_score()

    elif choice == "2":
        view_scores()

    elif choice == "3":
        delete_score()

    elif choice == "4":
        modify_score()

    elif choice == "5":
        clear_scores()

    elif choice == "6":
        save_scores()
        print("High scores saved to highscores.txt")
        print("Program exited.")

    else:
        print("Invalid choice. Please enter a number from 1 to 6.")
```
